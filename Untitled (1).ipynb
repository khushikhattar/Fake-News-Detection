```python
# Importing Libraries
!pip install pandas numpy seaborn scikit-learn nltk tpot
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import re
import string
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
import nltk
```

```python
# Downloading stopwords
nltk.download('stopwords')
```

```python
# Importing Dataset
df_fake = pd.read_csv("Fake.csv")
df_true = pd.read_csv("True.csv")
df_fake.head()
df_true.head()
```

```python
# Inserting a column "class" as target feature
df_fake["class"] = 0
df_true["class"] = 1
df_fake.shape, df_true.shape
```

```python
# Merging True and Fake Dataframes
df_merge = pd.concat([df_fake, df_true], axis=0)
```

```python
# Removing unnecessary columns
df = df_merge.drop(["title", "subject", "date"], axis=1)
```

```python
#Random Shuffling the dataframe
df = df.sample(frac = 1)
df.head()
df.reset_index(inplace = True)
df.drop(["index"], axis = 1, inplace = True)
df.columns
df.head()
```

```python
# Creating a function to process the texts
def wordopt(text):
    text = text.lower()
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r"https?://\S+|www\.\S+", '', text)
    text = re.sub(r'<.*?>+', '', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub(r'\n', '', text)
    text = re.sub(r'\w*\d\w*', '', text)    
    return text

```

```python
# Text processing
df["text"] = df["text"].apply(wordopt)
```

```python
# Defining dependent and independent variables
X = df["text"]
Y = df["class"]
```

```python
# Splitting Training and Testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)

# Converting Text to Vectors
vectorization = TfidfVectorizer()
Xv_train = vectorization.fit_transform(X_train)
Xv_test = vectorization.transform(X_test)
```

```python
# Function to plot confusion matrix
def plot_confusion_matrix(model_name, Y_test, Y_pred):
    cm = confusion_matrix(Y_test, Y_pred)
    plt.figure(figsize=(8, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix for {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()
```

```python
# Function to evaluate models
def evaluate_model(model, model_name):
    model.fit(Xv_train, Y_train)
    Y_pred = model.predict(Xv_test)

    accuracy = accuracy_score(Y_test, Y_pred)
    print(f"{model_name} Accuracy: {accuracy * 100:.2f}%")
    print(classification_report(Y_test, Y_pred))
    plot_confusion_matrix(model_name, Y_test, Y_pred)
    
    return Y_pred
```

```python
# Define models
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(random_state=0),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Naive Bayes": MultinomialNB(),
}

# Store predictions for comparison
predictions = {}
```

```python
# Evaluate each model
for model_name, model in models.items():
    print(f"\nEvaluating {model_name}...")
    Y_pred = evaluate_model(model, model_name)
    predictions[model_name] = Y_pred
```

```python
# Create a summary of the results
summary = pd.DataFrame({
    'Model': [],
    'Accuracy': [],
    'Precision': [],
    'Recall': [],
    'F1-Score': []
})

for model_name, model in models.items():
    Y_pred = predictions[model_name]
    report = classification_report(Y_test, Y_pred, output_dict=True)
    
    # Creating a new DataFrame for each model's performance
    model_performance = pd.DataFrame([{
        'Model': model_name,
        'Accuracy': accuracy_score(Y_test, Y_pred) * 100,
        'Precision': report['macro avg']['precision'] * 100,
        'Recall': report['macro avg']['recall'] * 100,
        'F1-Score': report['macro avg']['f1-score'] * 100
    }])
    
    # Concatenating the new DataFrame with the summary DataFrame
    summary = pd.concat([summary, model_performance], ignore_index=True)

print("\nSummary of Model Performance:")
print(summary)
```

```python
# Function to return a label
def output_label(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"

# Function for manual testing of individual news articles from CSV file
def manual_testing(csv_file):
    # Load the testing data from CSV file
    testing_data = pd.read_csv(csv_file)
    
    # Ensure the CSV has a column named "text"
    if "text" not in testing_data.columns:
        print("The CSV file should have a 'text' column.")
        return

    # Apply text processing
    testing_data["text"] = testing_data["text"].apply(wordopt)
    
    # Transform the text to feature vectors
    new_xv_test = vectorization.transform(testing_data["text"])
    
    # Make predictions using each model
    for model_name, model in models.items():
        predictions = model.predict(new_xv_test)
        print(f"\n{model_name} Predictions:")
        for idx, pred in enumerate(predictions):
            print(f"Article {idx+1}: {output_label(pred)}")

# Example usage for manual testing from CSV file
csv_file = "manual_testing.csv"  # Path to the CSV file
manual_testing(csv_file)
```

```python
# Import joblib for saving models
import joblib
import os

# Create a directory for saving models and vectorizer
os.makedirs("web_app_models", exist_ok=True)

# Save each trained model and the TF-IDF vectorizer
for model_name, model in models.items():
    model_filename = f"web_app_models/{model_name.replace(' ', '_').lower()}.joblib"
    joblib.dump(model, model_filename)
    print(f"{model_name} saved as {model_filename}")

# Save the vectorizer
vectorizer_path = "web_app_models/tfidf_vectorizer.joblib"
joblib.dump(vectorization, vectorizer_path)
print(f"TF-IDF Vectorizer saved as {vectorizer_path}")

# Save the summary of model performance in a CSV file
summary_csv_path = "web_app_models/model_performance_summary.csv"
summary.to_csv(summary_csv_path, index=False)
print(f"Model performance summary saved as {summary_csv_path}")
```
